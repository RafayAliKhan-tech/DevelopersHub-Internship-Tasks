Introduction

This task explores the Iris dataset to understand its structure and visualize relationships between variables using Python libraries.

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Dataset direct URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# 2. Column names define since this is not in raw file
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
# 3. Data load using pandas
df = pd.read_csv(url, names=column_names)
# 4. Validation
print("--- first 5 Rows ---")
print(df.head())

print("\n--- Data Summary ---")
print(df.info())
```

Step4: Scatter Plot (Relationship Check)

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="sepal_length", y="petal_length", hue="class", data=df)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

Step5: Histogram (Distribution Check)

```python
plt.figure(figsize=(6,4))
sns.histplot(df["sepal_length"], bins=20, kde=True)
plt.title("Distribution of Sepal Length")
plt.show()
```

Step6: Box Plot (Outliers Check)

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="class", y="petal_length", data=df)
plt.title("Petal Length by Classes")
plt.show()
```
